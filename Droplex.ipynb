#**Droplex**
- ddPCR 플랫폼 암 동반진단 검사
- 동반진단에 최적화된 **ddPCR 플랫폼**을 기반으로 환자의 혈액 및 조직으로부터 추출한 유전자를 분석하여 돌연변이를 검출
- **다양한 암종**에 대하여 **특정 치료제**의 효과를 보이는 환자들 선별


##**ddPCR 플랫폼**

- 기존 암 돌연변이 검출을 위한 조직 검사와 비교하여 **혈액 샘플은 DNA 추출량이 적기 때문**에 민감한 검사 필요
- 최첨단 Digital PCR 기술인 ddPCR은 **적은양의 DNA로도 정확한 돌연변이 검출이 가능**하여 액체 생검에 특화된 플랫폼
