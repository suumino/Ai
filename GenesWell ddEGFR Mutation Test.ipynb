#**GenesWell ddEGFR Mutation Test(폐암 동반진단)**

##**동반진단**
- 항암제는 **암 세포와 정상 세포 구분 없이** 빠르게 성장하는 세포들 공격 -> 백혈구 감소, 탈모, 구토 등의 **부작용** 수반
- 표적항암제는 암 세포에만 존재하는 **특정 표적**을 감지하여 공격 -> 부작용 적다
- 표적항암제는 특정 암세포에만 작용하여 사용범위가 제한적 -> 처방 전 **환자의 유전적 정보를 확인하는 동반진단검사**가 선행되어야함
- 동반진단검사 : 환자의 **유전체 정보를 파악**하여 특정 치료제에 효과를 보이거나 심각한 이상반응의 위험성으로부터 치료를 받을 수 없는 **환자들을 선별**하는 첨단 진단 검사



##**ddPCR(Droplet Digital PCR)**
- **ddPCR** : 2만개의 droplet으로 쪼개어 증폭 반응 시킨 후 타겟 유전물질을 계수하여 분석
- 젠큐릭스의 동반진단검사는 ddPCR기법을 적용하여 기존 PCR 기반 제품에 비해 우수함 ㅎㅎ

##**GenesWell ddEGFR Mutation Test**

- 비소세포폐암(NSCLC, Non-Small Cell Lung Cancer)환자를 대상으로 **45개 EGFR(Epidermal Growth Factor Receptor)돌연변이 존재 유무**를 알려주는 검사
- 검사 결과를 통해 비소세포폐암 **표적 치료제인 Tarceva(Erlotinib), Tagrisso(Osimertinib)투여 환자 선별**을 도울 수 있음
- 비소세포폐암에서 **EGFR 돌연변이**는 **EGFR-TKI로 사용되는 약물**에 대한 반응을 예측할 수 있는 **특정 표지자**
